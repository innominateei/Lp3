```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://github.com/afnan47/sem7/raw/main/ML/1.%20Uber%20Price%20Detection/uber.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.boxplot(df["fare_amount"])
```

```python
q_low = df["fare_amount"].quantile(0.01)
q_hi = df["fare_amount"].quantile(0.99)
df = df[(df["fare_amount"] < q_hi) & (df['fare_amount'] > q_low)]
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df["fare_amount"]
x = df.drop("fare_amount", axis = 1)
```

```python
x["pickup_datetime"] = pd.to_numeric(pd.to_datetime(x["pickup_datetime"]))
```

```python
x = x.loc[:,x.columns.str.contains('^Unnamed')]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 2529)
```

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(x_train,y_train)
```

```python
pred = Lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
Lrrmse = mean_squared_error(y_test,pred)
print("RMSE value:",Lrrmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
Rf = RandomForestRegressor(n_estimators=100, random_state=2529)
Rf.fit(x_train,y_train)
pred = Rf.predict(x_test)
from sklearn.metrics import mean_squared_error
Rfrmse =  np.sqrt(mean_squared_error(y_test,pred))
print("RMSE value:",Rfrmse)
```
